```python
import matplotlib.pyplot as plt
import numpy as np
from mplsoccer import Pitch, Sbopen, VerticalPitch
```

```python
# Opening the data set

parser = Sbopen()
idEvent = 69301 # indicates the event of that we are taking the data
df , related , freeze, tactics = parser.event(idEvent)
```

```python
#print(df.columns)

team1 , team2 = df.team_name.unique()
print('team 1 is ', team1, 'team 2 is ', team2 )
```

```python
def teamPasses(team):
    mask = (df.team_name==team)&(df.type_name=='Pass')
    df_reduced = df.loc[mask,['x', 'y', 'player_name']]
    return df_reduced

SwedenPasses = teamPasses(team2)
SwedenPasses.head()
```

```python
pitch= Pitch(line_color = 'black')
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)


def plotAllPasses(dataFrame , colorEntry):
    
    for i, row in dataFrame.iterrows():
        pitch.scatter(row.x, row.y , alpha = 0.6 , s= 500 , color = colorEntry, ax = ax['pitch'])

plotAllPasses(SwedenPasses,'blue')
```
